# Simulating Fermentation with iJO1366

## 1. Load the python packages

```python
import pandas as pd
import pandas
import cobra
pd.set_option('display.max_columns', 100)
pd.set_option('display.width',100)
pd.set_option('display.max_colwidth',300)
```

## 2. Load the iJO1366 model

Load the iJO1366 model in pickle format that is located in the current directory.

```python
import cobra
from cobra.io import load_model
model = load_model('iJO1366')
model_save =  model.copy()
```

## 3.0 Simulate Fermentation

```python
x = range(0, 30, 2)
x
```

```python
o2 = [0, -2, -4, -6, -8, -10, -15, -20]
growth_rate = []
acetate = []
ethanol = []
formate = []
acald = []
protons = []
co2 = []
glucose = []
oxygen = []
lactate = []

#model.reactions.get_by_id('EX_glc__D_e').lower_bound = -10
#model.reactions.get_by_id('EX_glc__D_e').upper_bound = -10

for x in range(0,-20,-2):
    #print('\nOxygen Flux # ',x)
    model.reactions.get_by_id('EX_o2_e').lower_bound = x
    model.reactions.get_by_id('EX_o2_e').upper_bound = x
    solution = model.optimize()
    growth_rate.append(solution.objective_value)
    acetate.append(model.reactions.EX_ac_e.x)
    ethanol.append(model.reactions.EX_etoh_e.x)
    formate.append(model.reactions.EX_for_e.x)
    acald.append(model.reactions.EX_acald_e.x)
    protons.append(model.reactions.EX_h_e.x)
    co2.append(model.reactions.EX_co2_e.x)
    glucose.append(model.reactions.EX_glc__D_e.x)
    oxygen.append(x)
    lactate.append(model.reactions.EX_lac__D_e.x)
    
growth_rate
```

```python
import matplotlib.pyplot as plt

# x axis values
x = oxygen
y = growth_rate
  
# plotting the points 
plt.plot(x, growth_rate, label = "Growth Rate")

# naming the x axis
plt.xlabel('Oxygen Flux')
# naming the y axis
plt.ylabel('Growth-rate')
  
# giving a title to my graph
plt.title('Growth-rate vs Oxygen Flux')
  
# function to show the plot
plt.show()
```

```python
# x axis values
x = range(0,-20,-2)
# corresponding y axis values
y = growth_rate

plt.figure(figsize=(12,8))

# plotting the points 
plt.plot(x, acetate, label = "acetate")
plt.plot(x, ethanol, label = "ethanol")
plt.plot(x, formate, label = "formate")
plt.plot(x, acald, label = "acald")
plt.plot(x, co2, label = "co2")
plt.plot(x, glucose, label = "glucose")
#plt.plot(x, protons, label = "protons")
plt.plot(x,lactate, label = "lactate")

# naming the x axis
plt.xlabel('Oxygen Flux')
# naming the y axis
plt.ylabel('Fermentation Flux Values')
  
# giving a title to my graph
plt.title('Fermentation Flux Values vs Oxygen Flux')
  
# function to show the plot
plt.legend()
#plt.yscale('log')
plt.show()
```

```python
model = model_save.copy()
growth_rate2 = []
acetate2 = []
ethanol2 = []
formate2 = []
acald2 = []
protons2 = []
co2_2 = []
glucose2 = []
oxygen2 = []
lactate2 = []

#model.reactions.get_by_id('EX_glc__D_e').lower_bound = -10
#model.reactions.get_by_id('EX_glc__D_e').upper_bound = -10

for y in range(0,-200,-1):
    #print('\nOxygen Flux # ',x)
    z = y/10
    model.reactions.get_by_id('EX_o2_e').lower_bound = z
    model.reactions.get_by_id('EX_o2_e').upper_bound = z
    solution2 = model.optimize()
    growth_rate2.append(solution2.objective_value)
    acetate2.append(model.reactions.EX_ac_e.x)
    ethanol2.append(model.reactions.EX_etoh_e.x)
    formate2.append(model.reactions.EX_for_e.x)
    acald2.append(model.reactions.EX_acald_e.x)
    protons2.append(model.reactions.EX_h_e.x)
    co2_2.append(model.reactions.EX_co2_e.x)
    glucose2.append(model.reactions.EX_glc__D_e.x)
    oxygen2.append(z)
    lactate2.append(model.reactions.EX_lac__D_e.x)
    
#growth_rate2
```

```python
import matplotlib.pyplot as plt

# x axis values
x = oxygen2
y = growth_rate2
  
# plotting the points 
plt.plot(x, growth_rate2, label = "Growth Rate")

# naming the x axis
plt.xlabel('Oxygen Flux')
# naming the y axis
plt.ylabel('Growth-rate')
  
# giving a title to my graph
plt.title('Growth-rate vs Oxygen Flux')
  
# function to show the plot
plt.show()
```

```python
# x axis values
xt = range(0,-200,-1)
x = [y/10 for y in xt ]

# corresponding y axis values
y = growth_rate2

plt.figure(figsize=(12,8))

# plotting the points 
plt.plot(x, acetate2, label = "acetate")
plt.plot(x, ethanol2, label = "ethanol")
plt.plot(x, formate2, label = "formate")
plt.plot(x, acald2, label = "acald")
plt.plot(x, co2_2, label = "co2")
plt.plot(x, glucose2, label = "glucose")
#plt.plot(x, protons2, label = "protons")
plt.plot(x,lactate2, label = "lactate")

# naming the x axis
plt.xlabel('Oxygen Flux')
# naming the y axis
plt.ylabel('Fermentation Flux Values')
  
# giving a title to my graph
plt.title('Fermentation Flux Values vs Oxygen Flux')
  
# function to show the plot
plt.legend()
#plt.yscale('log')
plt.show()
```
